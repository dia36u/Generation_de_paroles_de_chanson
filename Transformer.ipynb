# Traduction

```python
translator = pipeline('translation_en_to_fr',model="t5-base")
outputs = translator(text_3, clean_up_tokenization_spaces=True, min_length=100)
print(outputs[0]['translation_text'])
```

# Text Classification

```python
classifier = pipeline("text-classification",model="camembert-base")
outputs = classifier(text)
pd.DataFrame(outputs)
```

# Named Entity Recognition

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger(text_3)
pd.DataFrame(outputs)
```

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger("Donald Trump, the former president, is now playing golf all days in Florida with his friend Bill")
pd.DataFrame(outputs)
```

# Question Answering

```python
reader = pipeline("question-answering")
results = []
questions = ["What is Quidditch?"]
for question in questions:
    outputs = reader(question=question, context=text_3)
    results.append(outputs)

pd.DataFrame(results)
```

# Summarization

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=150, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

# Text Generation

```python
generator = pipeline("text-generation")
response = "Dear friends. Let me tell you my last experience with this online store where I ordered a book."
prompt = text + "\n\nStory I told to my friends:\n\n" + response
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])
```
