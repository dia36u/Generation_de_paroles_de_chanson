```python
# importing the dataset
import pandas as pd
import numpy
from sklearn import preprocessing
```

```python
df_artists = pd.read_csv(r'dataset\artists-data.csv')
df_lyrics = pd.read_csv(r'dataset\lyrics-data.csv')
```

```python
df=pd.merge(df_artists,df_lyrics,left_on='Link',right_on='ALink')
df_genre=df['Genres'].str.split(';',expand=True)
for colums in df_genre:
    df_genre[colums] = df_genre[colums].str.strip()
df=pd.merge(df,df_genre,left_on=df.index,right_on=df_genre.index)
df.rename(columns={0: "Genre 1", 1: "Genre 2",2: "Genre 3", 3: "Genre 4"}, inplace=True)
df.drop(['SLink', 'key_0','Genres','Link','ALink'], axis=1,inplace=True)
```

```python
from collections import Counter

def genre_by_language(language, df=df):
    genre={}
    df_lang = df[df['language']==language]
    dict_0 = df_lang["Genre 1"].value_counts().to_dict()
    dict_1 = df_lang["Genre 2"].value_counts().to_dict()
    dict_2 = df_lang["Genre 3"].value_counts().to_dict()
    dict_3 = df_lang["Genre 4"].value_counts().to_dict()
    genre = Counter(dict_0) + Counter(dict_1) + Counter(dict_2) + Counter(dict_3)
    return(genre)
```

```python
genre=genre_by_language("en")
sorted(genre.items(), key=lambda t: t[1], reverse=True)
```

```python
first_genre=genre="Rock"
first_lg="en"
```

```python
df=df.iloc[df.index[(df['Genre 1']==first_genre) | (df['Genre 2']==first_genre) | (df['Genre 3']==first_genre) | (df['Genre 4']==first_genre)].tolist()]
df=df[df['language']==first_lg]
df.drop(['language', 'Genre 1','Genre 2','Genre 3','Genre 4'], axis=1,inplace=True)
df
```
